```python
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy import interpolate
import time
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
#this next line is only needed in iPython notebooks
%matplotlib inline 
import math
import matplotlib.font_manager as fm
import matplotlib.ticker as mtick
font = fm.FontProperties(size = 12)
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
def hide_spines(intx=False,inty=False):
    """Hides the top and rightmost axis spines from view for all active
    figures and their respective axes."""

    # Retrieve a list of all current figures.
    figures = [x for x in matplotlib._pylab_helpers.Gcf.get_all_fig_managers()]
    if (plt.gca().get_legend()):
        plt.setp(plt.gca().get_legend().get_texts(), fontproperties=font) 
    for figure in figures:
        # Get all Axis instances related to the figure.
        for ax in figure.canvas.figure.get_axes():
            # Disable spines.
            ax.spines['right'].set_color('none')
            ax.spines['top'].set_color('none')
            # Disable ticks.
            ax.xaxis.set_ticks_position('bottom')
            ax.yaxis.set_ticks_position('left')
           # ax.xaxis.set_major_formatter(mtick.FuncFormatter(lambda v,_: ("10$^{%d}$" % math.log(v,10)) ))
            for label in ax.get_xticklabels() :
                label.set_fontproperties(font)
            for label in ax.get_yticklabels() :
                label.set_fontproperties(font)
            #ax.set_xticklabels(ax.get_xticks(), fontproperties = font)
            ax.set_xlabel(ax.get_xlabel(), fontproperties = font)
            ax.set_ylabel(ax.get_ylabel(), fontproperties = font)
            ax.set_title(ax.get_title(), fontproperties = font)
            if (inty):
                ax.yaxis.set_major_formatter(mtick.FormatStrFormatter('%d'))
            if (intx):
                ax.xaxis.set_major_formatter(mtick.FormatStrFormatter('%d'))
def show(nm,a=0,b=0):
    hide_spines(a,b)
    #ax.xaxis.set_major_formatter(mtick.FuncFormatter(lambda v,_: ("10$^{%d}$" % math.log(v,10)) ))
    #plt.yticks([1,1e-2,1e-4,1e-6,1e-8,1e-10,1e-12], labels)
    #ax.yaxis.set_major_formatter(mtick.FuncFormatter(lambda v,_: ("10$^{%d}$" % math.log(v,10)) ))
    if (len(nm)>0):
        plt.savefig(nm+".pdf",bbox_inches='tight');
    plt.show()
    
    
import random

%matplotlib inline
```

```python
np.random.seed(42)
random.seed(42)
mean_each = 50
N1 = int(np.random.normal(loc=mean_each,scale=4,size=1))
N2 = int(np.random.normal(loc=mean_each,scale=4,size=1))
N3 = int(np.random.normal(loc=mean_each,scale=4,size=1))

print(N1,N2,N3)
```

```python
X = np.zeros((N1+N2+N3,2))
rval = np.zeros(N1+N2+N3,dtype=int) + 0
rval[0:N1] = 2
rval[N1:(N1+N2)] = 1
np.random.seed(42)
X[0:N1,0] = np.random.normal(loc=np.repeat(1,N1),scale=np.repeat(0.75,N1))
np.random.seed(43)
X[0:N1,1] = np.random.normal(loc=np.repeat(0,N1),scale=np.repeat(0.75,N1))

np.random.seed(402)
X[N1:(N1+N2),0] = np.random.normal(loc=np.repeat(1.75,N2),scale=np.repeat(.75,N2))
np.random.seed(403)
X[N1:(N1+N2),1] = np.random.normal(loc=np.repeat(2,N2),scale=np.repeat(.75,N2))

np.random.seed(4002)
X[(N1+N2):(N1+N2+N3),0] = np.random.normal(loc=np.repeat(4,N3),scale=np.repeat(.75,N3))
np.random.seed(4003)
X[(N1+N2):(N1+N2+N3),1] = np.random.normal(loc=np.repeat(0.5,N3),scale=np.repeat(.75,N3))
```

```python
plt.plot(X[(N1+N2):(N1+N2+N3),0],X[(N1+N2):(N1+N2+N3),1],"^")
plt.plot(X[N1:(N1+N2),0],X[N1:(N1+N2),1],"o")
plt.plot(X[0:N1,0],X[0:N1,1],"*")
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
show("clusts_close")
```

```python
from sklearn import metrics
from sklearn.cluster import KMeans
```

```python

kmeans = KMeans(n_clusters=3, random_state=3).fit(X)
print(kmeans.cluster_centers_)
pred= kmeans.predict(X)
m = ['^', 'o','*']
fig = plt.gcf()
ax = fig.gca()
for i in range(3):
    plt.scatter(X[(pred==i)*(rval==i),0], X[(pred==i)*(rval==i),1], marker=m[i])
    plt.scatter(X[(pred==i)*(rval!=i),0], X[(pred==i)*(rval!=i),1], marker=m[i], facecolors='none', edgecolors='r',s=60)
    plt.plot(kmeans.cluster_centers_[i][0],kmeans.cluster_centers_[i][1],'k',marker=m[i],markersize=15,alpha=0.5)

    #r1 = np.max([(X[pred==i,0]-kmeans.cluster_centers_[i][0])**2 + (X[pred==i,1]-kmeans.cluster_centers_[i][1])**2])
    #circle1 = plt.Circle(kmeans.cluster_centers_[i], math.sqrt(r1), color='r',fill=None)
    #ax.add_artist(circle1)

plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
print(kmeans.inertia_,kmeans.inertia_*3)
show("clusts_close_kmeans")
Xclose = X.copy()

```

```python
Nk = 20
LK = np.zeros((Nk,2))
for k in range(Nk):
    kmeans = KMeans(n_clusters=k+1, random_state=3).fit(X)
    LK[k,0] = kmeans.inertia_*(k+1)
    kmeans = KMeans(n_clusters=k+1, random_state=3).fit(Xclose)
    LK[k,1] = kmeans.inertia_*(k+1)
plt.plot(range(1,Nk+1),LK[:,0],'o-',label="Std.=0.6")
plt.plot(range(1,Nk+1),LK[:,1],'*-',label="Std.=0.75")
plt.xticks(range(1,Nk+1,2))
plt.legend(loc="best")
plt.xlabel("$K$")
plt.ylabel("LK")
show("Kchoose")
```
